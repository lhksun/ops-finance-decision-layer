```sql
CREATE OR ALTER VIEW dbo.vw__CM__STIP_x_CPDash_current
AS
SELECT
    /* ========== join grain columns (add period) ========== */
    s.FiscalYearPeriodDate,
    s.CurrSoldToSalesOfficeCd,
    s.CurrMatGrpCd,
    s.CurrProdDivCd,

    /* ========== STIP columns (requested) ========== */
    s.Manufacturer,
    s.SalesTypeCd,
    s.GrossSales,
    s.CostOfGoodsSold,
    s.GroupRebateAccrualAmount,
    s.CorporateRebateAccrualAmount,
    s.CustomerIncentiveRebateAmt,
    s.MaterialGroupDescription,
    s.ProductDivisionCode,
    s.ProductDivisionDescription,

    /* ========== CPDash KPI columns ========== */
    c.[PPDWriteOff%],
    c.[COGSManAdj%],
    c.[(Over)/UnderAbs%],
    c.[OutboundFreight%]

FROM dbo.vw__CM__STIP_summary_current s
LEFT JOIN dbo.vw__CM__CPDash_current c
    ON  c.CurrMatGrpCd = s.CurrMatGrpCd
    AND c.CurrSoldToSalesOfficeCd = s.CurrSoldToSalesOfficeCd
    AND c.FiscalYearPeriodDate = s.FiscalYearPeriodDate
    AND c.CurrProdDivCd = s.CurrProdDivCd;
GO
```

```sql
SELECT TOP (20) *
FROM dbo.vw__CM__STIP_x_CPDash_current
ORDER BY FiscalYearPeriodDate DESC;
```

```sql
SELECT TOP (50)
  CurrSoldToSalesOfficeCd,
  FiscalYearPeriodDate,
  CurrMatGrpCd,
  CurrProdDivCd,
  COUNT(*) AS RowCnt
FROM dbo.vw__CM__STIP_x_CPDash_current
GROUP BY
  CurrSoldToSalesOfficeCd,
  FiscalYearPeriodDate,
  CurrMatGrpCd,
  CurrProdDivCd
HAVING COUNT(*) > 1
ORDER BY RowCnt DESC;
```
